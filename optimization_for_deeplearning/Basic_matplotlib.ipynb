```python
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

## 첫 번째 코드는 주피터 노트북에서 그림이 나타나도록 하는 코드
## 두 번째 코드는 그래프가 고해상도로 그려질 수 있게 설정해주는 코드
```

```python
## 분포도(Scatter plot) 그리기
import numpy as np
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)

plt.scatter(x, y)
plt.show()
```

```python
## 위 분포도에서는 좌표 외에 어떤 정보도 존재하지 않는다
## 점이 찍히는 방식 대신 원을 그리는 방식으로 접근해보면?
## 원의 크기에도 의미를 부여할 수 있다
## plt.scatter(x, y, s=area)와 같이
## s에 해당 정보(area)를 넣어주면 다음 그림과 같이
## 각기 다른 크기의 원이 그려지는 것을 확인할 수 있다
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area)
plt.show()
```

```python
## 투명도 조절하기
## 0 < alpha < 1
## 0은 완전히 투명한 상태, 1은 완전히 불투명한 상태
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area, alpha=0.5)
plt.show()
```

```python
## 분포도에 색깔 적용하기
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
## 페어플랏(pair plot)
## 모든 변수에 대해서 분포도와 히스토그램을 그려주는 기능이
## seaborn 패키지에 있다
import seaborn as sns
sns.set(style='ticks', color_codes=True)
iris = sns.load_dataset('iris')
g = sns.pairplot(iris)
```

```python
## 페어플랏은 범주형 변수에 따라서 색깔과 마커의 모양을 다르게 표시해줄 수 있다
g = sns.pairplot(iris, hue='species', markers=['o', 's', 'd'])
```

```python
## 단일 변수 함수 그래프 그리기
## y = x^2의 그래프를 [0,5]에서 그리고 싶을 때
# 정의역(domain)을 선언한다, x = np.arange(0.,5.,0.2)
# 선언된 정의역으로 함숫값을 계산한다, y = x**2
# plot(x, y)를 호출한다
# show()를 호출하여 그래프를 화면에 띄운다
x = np.arange(0.,5.,0.2)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0.,5.,1.125)
y = x**2
plt.plot(x, y)
plt.show()

## 곡선의 곡률이 큰 함수일 수록 간격을 좁게 설정해줘야 왜곡 없는 그래프를 얻을 수 있다
```

```python
## 여러 그래프를 한 눈에 보기
# 다음 세 가지 그래프를 같은 축에서 공유하도록 그리는 코드
# y = x
# y = 1/4 * x^2
# y = cos(x)

## plot()은 네 가지 선 모양을 제공한다
# '-' : solid
# '--' : dashed
# '-.' : dash-dot
# ':' : dotted

x = np.arange(0., 5., 0.2)
y1 = x
y2 = 0.25 * x**2
y3 = np.cos(x)
plt.plot(x, y1, '-')
plt.plot(x, y2, '--')
plt.plot(x, y3, ':')
plt.show()
```

```python
## 점들의 위치를 마커로 표시할 수 있는 기능도 있다
## 선 모양과 함께 세 번째 인자로 넣어주면 된다
x = np.arange(0., 5., 0.2)
y1 = x
y2 = 0.25 * x**2
y3 = np.cos(x)
plt.plot(x, y1, '-o')
plt.plot(x, y2, '--s')
plt.plot(x, y3, ':x')
plt.show()
```

```python
## 각 그래프 이름 달아주기; 범례
## plt.legend()
x = np.arange(0., 5., 0.2)
y1 = x
y2 = 0.25 * x**2
y3 = np.cos(x)
plt.plot(x, y1, '-o')
plt.plot(x, y2, '--s')
plt.plot(x, y3, ':x')
plt.legend(['y=x', 'y=0.25 * x**2', 'y = cos(x)'])
plt.show()
```

```python
## 그래프 스타일링
## 제목 : plt.title()
## 눈금 : plt.grid()
## x, y축 이름 : plt.xlabel(), plt.ylabel()
x = np.arange(0., 5., 0.2)
y1 = x
y2 = 0.25 * x**2
y3 = np.cos(x)
plt.plot(x, y1, '-o')
plt.plot(x, y2, '--s')
plt.plot(x, y3, ':x')
plt.title('Plot tutorial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['y=x', 'y=0.25 * x**2', 'y = cos(x)'])

plt.show()
```

```python
## 매번 plot() 범례에 나올 이름을 지정하는 방법도 있음
## legend()에 입력할 인자는 필요하지 않다
x = np.arange(0., 5., 0.2)
y1 = x
y2 = 0.25 * x**2
y3 = np.cos(x)
plt.plot(x, y1, '-o', label=r'$y=x$')
plt.plot(x, y2, '--s', label=r'$y=\frac{1}{4}x^2$')
plt.plot(x, y3, ':x', label=r'$y=cos(x)$')
plt.title('Plot tutorial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

plt.show()
```

```python
## 세 개의 그래프 각각 따로 나타내기; subplot()
## plt.subplot(3,1,1)은 3개의 행과 1개의 열의 첫 번째 그래프
plt.subplot(3,1,1)
plt.plot(x, y1, '-ok')
plt.title('Subplot Example')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,1,2)
plt.plot(x, y2, '--sk')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,1,3)
plt.plot(x, y3, ':xk')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
```

```python
## 다변수 함수 그래프 그리기
## 다음 코드는 정사각형 정의역([-5,5]x[-5,5])에서
## z = sin(root(x^2 + y^2))을 그리는 코드
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

## 컨투어(Contour)는 같은 함숫값(z)를 갖는 선을 이어주는 그래프, 등고선과 같은 개념
fig = plt.figure(figsize=(5,5))
plt.contour(X, Y, Z, cmap='coolwarm')
plt.grid()
plt.axis('equal')
plt.show()
```

```python
## 컨투어를 구성하는 선들 사이를 해당 색깔로 채운 그래프는 contourf()를 사용
## 기존 contour() 명령어 뒤에 f가 붙어있는데 이것은 Filled의 약자
fig = plt.figure(figsize=(5,5))
plt.contourf(X, Y, Z, cmap='coolwarm')
plt.grid()
plt.axis('equal')
plt.show()
```
